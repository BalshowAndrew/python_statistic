```python
# замена 'tbd' на NaN
df['User_Score'] = df['User_Score'].replace('tbd', np.NaN)
df['User_Score'].unique()
```

```python
# исправим типы данных
df['Year_of_Release'] = df['Year_of_Release'].astype('Int64')
df['User_Count'] = df['User_Count'].astype('Int64')
df['User_Score'] = df['User_Score'].astype('float64')
df['Critic_Count'] = df['Critic_Count'].astype('Int64')
```

```python
# переводим оценки в один формат
df['User_Score'] = df['User_Score'] * 10
```

## График
Строим график динамики продаж видеоигр

```python
dynamics_by_region = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Year_of_Release']] \
        .groupby('Year_of_Release').sum()
dynamics_by_region.head()
```

```python
plt.plot(dynamics_by_region.index, dynamics_by_region['Global_Sales'])
plt.title('Динамика продаж видеоигр')
plt.xlabel('Год')
plt.ylabel('Продажи, млн.')
plt.savefig('11.png')
plt.show()
```

```python
# используем цикл для построения графиков продаж для каждого региона
for region in dynamics_by_region.columns:
    plt.plot(dynamics_by_region.index, dynamics_by_region[region])
plt.title('Динамика продаж видеоигр')
plt.xlabel('Год')
plt.ylabel('Продажиб млн.')
plt.legend(dynamics_by_region)
plt.show()
```

Метод .plot() в pandas основан на matplotlib. Если мы работаем с датафреймом, то это, как правило, более удобный вариант, чем использование matplotlib самой по себе.

Если работаем с табличными структурами, то можем воспользоваться функционалом, встроенным в pandas:
- метод plot по умолчанию использует индекс датафрейма для оси X
- метод plot по умолчанию строит график

```python
dynamics_by_region['Global_Sales'].plot(title='Динамика продаж видеоигр',
                                       xlabel='Год',
                                       ylabel='Продажиб млн.')
```

- plot без цикла построит графики для каждого региона (столбца)
- если у plot нет нужных методов, можно применит функции matplotlib

```python
dynamics_by_region.plot(title='Динамика продаж видеоигр', 
                       xlabel='Год',
                       ylabel='Продажи, млн.')
# следующие функции из matplotlib
plt.text(2007, 680, 'Почему отсуда начался спад?')
plt.legend(loc='upper left')
```

## Area chart

## Столбчатая диаграмма (Bar chart)

```python
# количество игр в разрезе платформ
games_by_platforms = df[['Name', 'Platform']].groupby('Platform').count()
games_by_platforms.head()
```

```python
# выберем платформы, по которым будем строить график
games_by_platforms.loc[['PC', 'X360', 'PS3', 'PS4', 'XOne']].sort_values(by='Name', ascending=False).plot(kind='bar',
                                                                                                         title='Топ 5 платформ по количеству игр',
                                                                                                         xlabel='Платформа',
                                                                                                         ylabel='Количество игр')
```

```python
# топ-5 жанров по продажам
top5_genres = df[['Global_Sales', 'Genre']].groupby('Genre').sum().sort_values('Global_Sales', ascending=False).head(5)
top5_genres.plot(kind='bar',
                title='Топ-5 жанров по продажам',
                xlabel='Платформа',
                ylabel='Количетво игр')
for i, val in enumerate(top5_genres.values):
    plt.text(i, val, int(val),
             horizontalalignment='center',
             verticalalignment='bottom',
             fontdict={'size':14}
            )
```

```python
# Диаграмма продаж по регионам (выглядит плохо)
dynamics_by_region.plot(kind='bar')
```

```python
# лучше сделать с накоплением
dynamics_by_region.drop('Global_Sales', axis=1).plot(kind='bar', stacked=True, figsize=(12, 4))
```
